```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, scale_fill_manual
import plotnine as p9

df_athlete=pd.read_csv('../../../datos/athlete_events.csv')
df_region=pd.read_csv('../../../datos/noc_regions.csv')

athlete_china=df_athlete[df_athlete['Team']=='China']

athlete_china.dropna(inplace=True)

# Convertir la columna a tipo categórico con el orden deseado
athlete_china['Medal']=pd.Categorical(athlete_china['Medal'], categories=['Bronze', 'Silver', 'Gold'], ordered=True)

athlete_china=athlete_china.sort_values(by='Medal').reset_index(drop=True)

bar_plot = (
    ggplot(athlete_china)
    + aes(x='factor(Sport)', fill='factor(Medal)')
    + geom_bar()
    + scale_fill_manual(values={'Gold': 'yellow', 'Silver': 'grey', 'Bronze': 'brown'})
    + p9.xlab('Deporte')
    + p9.ylab('Numero de medallas')
    
)
bar_plot.show()

#funcion agg
```
