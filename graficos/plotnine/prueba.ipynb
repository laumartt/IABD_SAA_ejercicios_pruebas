```python
from plotnine import *
```

```python
from plotnine import ggplot, geom_point, aes
from plotnine.data import mtcars

print("Dades `mtcars`:")
print(mtcars)

plot = (
    ggplot(mtcars)
    + aes("wt", "mpg", color="factor(gear)")
    + geom_point()
    + labs(title="Consum (wt) vs Pes (mpg)",
           x="Pes en lliures",
           y="Consum en milles per gal",
           color="Nombre de marxes"
    )
    + theme_classic()
)
'''
aes(x, y, **kwargs)
    color: color dels elements
    fill: color de fons dels elements

geom_point: punts.
geom_line: lnies.
geom_bar: barres.
geom_boxplot: diagrama de caixes.'''
plot.show()


'''labs: Permet especificar el ttol i les etiquetes dels eixos X i Y i altres elements.
xlab: etiqueta de l'eix X.
ylab: etiqueta de l'eix Y.'''

plot = (
    ggplot(mtcars)
    + aes("wt", "mpg", color="factor(gear)")
    + geom_line()
)

plot.show()


bar_plot = (
    ggplot(mtcars)
    + aes(x='factor(cyl)', fill='factor(gear)')
    + geom_bar()
)

bar_plot.show()
```
